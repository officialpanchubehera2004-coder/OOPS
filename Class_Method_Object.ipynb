```python
class computer():        
     
    def __init__(self,cpu,ram): # cpu & ram as argument
        self.cpu = cpu #hear self is the object & we need to refer to the object
        self.ram = ram 

    def config(self):     
        print('config is', self.cpu, self.ram)

com1= computer('i5','8gb')  
com2= computer('i7','32gb')

com1.config()
com2.config()
```

```python
class computer():        
     
    def __init__(self,cpu,ram,price ): # cpu & ram as argument
        self.cpu = cpu #hear self is the object & we need to refer to the object
        self.ram = ram 
        self.price = price
 
    def config(self):     
        print('config is', self.cpu, self.ram, self.price ) 

com1 = computer('i5', 8, 1000)
com2 = computer('i7', 16, 2000)

com1.config() 
com2.config()
```

```python
class computer():        
     
    def __init__(self, cpu, ram, price, company ): # cpu & ram as argument
        
        self.cpu = cpu #hear self is the object & we need to refer to the object
        self.ram = ram 
        self.price = price
        self.company= company
 
    def config(self):     
        print('config is', self.cpu, self.ram, self.price , self.company) 

com1 = computer('i5', 8, 1000, 'len' )
com2 = computer('i7', 16, 2000,'hp')

com1.config() 
com2.config()
```
